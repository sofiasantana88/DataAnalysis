```python
import pymongo
import certifi

client = pymongo.MongoClient(secret_key, tlsCAFile=certifi.where())
print(f"Using MongoDB version {client.server_info()['version']}.")

my_database = client['DATA320']

print(f"This MongoDB server has the databases {my_database}")

all_collections = my_database.list_collection_names()

print(f"This database has the collections{all_collections}")
```

```python
import pandas as pd 

cursor = my_database['IMBD'].find()

imbd = pd.DataFrame(cursor)

imbd.head()
```

```python
imbd.release_date = pd.to_datetime(imbd.release_date, errors='coerce')
imbd.budget = pd.to_numeric(imbd.budget, errors='coerce')
imbd.runtime = pd.to_numeric(imbd.runtime, errors='coerce')
```

```python
# I chose to do the year 2017

movies2017 = imbd[imbd['release_date'].dt.year == 2017]

movies2017.head()
```

```python
# Now for the highest, and lowest lowest budget in 2017

lowest_budget2017 = movies2017.sort_values(by=['budget']).head()
lowest_budget2017
```

```python
# Now for the highest, and lowest lowest budget in 2017

highest_budget = movies2017.sort_values(by='budget', ascending=False).head()

highest_budget
```

```python
lowest_runtime = movies2017.sort_values(by='runtime').head()

lowest_runtime
```

```python
highest_runtime = movies2017.sort_values(by='runtime', ascending=False).head()

highest_runtime
```

I decided to investigate the year 2017 for this database. 

One of the first outliers I noticed was in the budget column. The movie Lover for a Day showed a budget of just 2.0, which seemed odd. On the other end, Oxidan had a listed budget of 2,000,000,000. That’s a huge difference, and it didn’t seem right. I figured they might be using different currencies, but the data doesn’t say which ones or a country for each movie, so I had to search for some of the outlierts to make sure.

I looked it up and found that IMDb actually lists Lover for a Day with a budget of 2 euros (https://www.imdb.com/es/title/tt5642124), so that number is actually technically correct. Another example of the currency problem is the movie Duvvada Jagannadham, which has a budget listed as 65.0, but after checking, it turns out the actual budget is 75 crores (Indian currency). So clearly, not all the movies are using the same units, which makes it hard to compare them properly. The same issue happens with Satan’s Slaves, listed in the higher budget number, that's because its budget is listed in rupiahs.

Now, regarding the runtime, the shortest movie listed is The Problem with Apu, with 49 minutes, which seemed very short, so I researched and all the movies that I found for 2017 were correct. 

For the movie Long Strange Trip, listed with a duration of 234 minutes is actually a miniseries of 6 episodes released in Amazon Prime in 2016, not a movie, this tells me that this database could have different types of entretainment (TV show, Movie, etc).
